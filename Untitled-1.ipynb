# IMDB top 250 rated movies

Today I will scrap IMBD website,  top 250 movies DATA into excel file.     


Format -     movie_rank /    movie_name /   release_year /    rating  

```python
from bs4 import BeautifulSoup                                         # install modules if not installed
import requests 
import openpyxl                                                       # openpyxl loads data into excel
excel = openpyxl.Workbook()
# print(excel.sheetnames)                                             # it returns [sheet], means it has only one sheet with name 'Sheet'
sheet = excel.active                                                  # Without .xlsx file will be text file
sheet.title = 'IMDB top 250 rated movies.xlsx'
print(excel.sheetnames)                                               # it returrns [IMDB top 250 rated movies]
sheet.append(['movie_rank','movie_name','release_year','rating'])     # created sheet with column names

try:
    source = requests.get('https://www.imdb.com/chart/top/')        
                                                                    # If URL has error it won't pop up
                                                                    # hence using alert 
    source.raise_for_status()                                       # Putting whole code in to 'Try-except' to prevent crashing
                                                                    # Source is my response object
    soup = BeautifulSoup(source.text, 'html.parser')                # source.text will extract HTML code of my response object
                                                                    # html.parser is default parser, it creates HTML tree
                                                                    # Beautifulsoup is funciton to use parser & create HTML tree
                                                                    # Soup will be mine Beautifulsoup object
    # print(soup) will print whole HTML code
    # go to website & R-click on text you want to extract, then click inspect to get HTML code linked to that element
    # note down it's tag
    # <tbody...>...</tbody> contains multiple elements <tr>...</tr> 
    # each having 5 <td>...</td> showing
    # postercolumn - titlecolumn - ratingcolumn_imbdrating - watchlistColumn 

    movies = soup.find('tbody' , class_= "lister-list").find_all('tr')  # Required information is under 
                                                                        # <tbody class="lister-list"> ...</tbody>
                                                                        # using just find() will extract only first <td> tag
                                                                        # using find_all will extract all tags names 'tr'
                                                                        # tr have no atribute in it
                                                                        # using find_all returns 'list'with length 250
    # print(len(movies)) gives 250
    for movie in movies :
        name = movie.find('td', class_="titleColumn").a.text # 'td' is name of tag with class = "titleColumn"
                                                             # keep printing every variable to know what they contain & keep going ahead
                                                             # 'a' is tag inside 'td' tag
                                                             # .text will give text between tags <td>...</td>
        
        rank = movie.find('td', class_="titleColumn").get_text(strip=True).split('.')[0]  
                                                                    # here (rank with .text) will return text of all tags inside it
                                                                    # rank with .get_text(strip=True) will return text of all tags 
                                                                    # without space, enter, tags
                                                                    # we just want rank & nothing else
                                                                    # we split string by '.'
                                                                    # .get_text(strip=True).split('.') this returns rank as LIST with 2 
                                                                    # elements
                                                                    # first is rank, secod is remanining text
                                                                    # ['1','The Shaw...'] 
                                                                    # [0] returns index 0 of list
        year = movie.find('td', class_="titleColumn").span.text.strip('()')
                                                                    # 'span' is tag inside 'td' tag 
                                                                    # strip('()') removes parenthesis from string
        rating = movie.find('td', class_="ratingColumn imdbRating").strong.text
                                                                    #'td' is name of tag with class ="ratingColumn imdRating" 
                                                                    # strong is another tag inside td
                                                                    # .text will return text between these tags 
        print(rank,"   ", name,"   ",year,"   ",rating)             # It prints whole list of movies
        
        sheet.append([rank,name,year,rating])                       # Saving data in excel
except Exception as e :
    print("Error in URL")

excel.save('IMDB top 250 rated movies.xlsx')                        # To save excel file
```
